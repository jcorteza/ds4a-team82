```python
dtypes={'LEAID': 'str'}
grades_data_sets = []
for x in range(10, 18):
    df = pd.read_csv(f'./data_sets/Assessments/math-achievement-lea-sy20{x}-{x + 1}.csv', dtype=dtypes)
    columns_of_interest = []
    for column in df.columns:
        if 'all_mth04' in column.lower() or column.lower() == 'stnam' or column.lower() == 'leaid' or 'leanm' in column.lower():
            columns_of_interest.append(column)

    df = df.loc[:, columns_of_interest]
    column_names = {}
    for column in columns_of_interest:
        if 'stnam' in column.lower():
            column_names[column] = 'STATE'
        elif 'leanm' in column.lower():
            column_names[column] = 'NAME'
        elif 'numvalid' in column.lower():
            column_names[column] = 'TOTAL_SCORES'
        elif 'pctprof' in column.lower():
            column_names[column] = 'PCT_PASS'
        elif 'leaid' in column.lower():
            column_names[column] = 'NCESID'

    df.rename(columns=column_names, inplace=True)
    df['YRDATA'] = int(x)
    grades_data_sets.append(df)

def format_state(state):
    words = state.split()
    new_name = state
    if  state.upper() == 'DISTRICT OF COLUMBIA':
        return 'DC'
    elif len(words) > 1:
        for i in range(len(words)):
            word = words[i]
            words[i] = word[0].upper() + word[1:].lower()
        new_name = ' '.join(words)
    else:
        new_name = new_name[0].upper() + new_name[1:].lower()
    
    if new_name in states.keys():
        return states[new_name]
    else:
        return new_name
        
with open('./data_sets/us_census/states_map.json') as f:
    states = json.load(f)
grades_data = pd.concat(grades_data_sets, sort=False)
grades_data['STATE'] = grades_data['STATE'].map(format_state)
grades_data['NAME'] = grades_data['NAME'].map(lambda x: x.upper())
grades_data['NCESID'] = grades_data['NCESID'].astype(str)
grades_data.head()
```

```python
grades_data.to_csv('./data_sets/Assessments/clean_grades_data.csv', na_rep='NA', index=False)
```
